### Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
y_train_pred = lr_model_16.predict(X_train_sm)

```

## Model Evaluation

Let's now plot the graph for actual versus predicted values.

```python
# Plotting y_test and y_pred to understand the spread

fig = plt.figure()
plt.scatter(y_test, y_pred_m16)
fig.suptitle('y_test vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('y_test', fontsize = 18)                          # X-label
plt.ylabel('y_pred', fontsize = 16)      
```

```python
#Regression plot
plt.figure(figsize=(15,8))
sns.regplot(x=y_test, y=y_pred_m16, ci=68, fit_reg=True,scatter_kws={"color": "blue"}, line_kws={"color": "red"})

plt.title('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)                          # Y-label
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_m16)
```

```python

print('The R-Squared score of the model for the predicted values is',round(r2_score(y_test, y_pred_m16),2))
```

We can see that the equation of our best fitted line is:

$ cnt = 0.2341  \times  yr + 0.4719  \times  temp + 0.0429 \times summer + 0.0783 \times winter + 0.0758 \times sep - 0.1138 \times holiday - 0.0221 \times workingday - 0.1550 \times windspeed - 0.0618 \times spring - 0.0377 \times Jan - 0.0513 \times July - 0.2873 \times Light Snow - 0.0825 \times Mist + Cloudy $


## Interpretation

* We can see that temperature variable is having the highest coefficient 0.4719, which means if the temperature increases by one unit the number of bike rentals increases by 0.4914 units.
* Since the bookings increase on good weather days with hotter temperature, the company must increase their bike availibilty and promotions during the summer months to further increase their booking count.

* We also see there are some variables with negative coefficients, A negative coefficient suggests that as the independent variable increases, the dependent variable tends to decrease. We have spring, mist cloudy , light snow variables with negative coefficient. 
* Can focus more on Summer & Winter season, September month,as they have good influence on bike rentals.
* we have got negative coefficients for Mist +cloudy and Lightsnow weather... And yes we can give offers
* An R-Squared value of 0.81 on the test data signifies that the model is a very good predictor and 81% of the variance is captured by the model.It can be further improved by using other regression techniques like Random Forest.
